```python
from sklearn.model_selection import train_test_split as tts
from sklearn.metrics import multilabel_confusion_matrix as mcm
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from IPython.display import display
from sklearn.linear_model import LogisticRegression as logit
from sklearn.ensemble import RandomForestClassifier as rfc
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier as knn
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

def kmeans(n):
  global y_class
  k_means = KMeans(n_clusters=n)
  k_means.fit(X)
  y_class = k_means.labels_

def pca(X):
  global pca_mod
  global Xpca
  #Componentes principales.
  pca_mod = PCA(n_components=2)
  #Matriz de autovectores
  pca_mod.fit(X)
  Xpca = pca_mod.transform(X)
  #Reducción DIFERENCIAR los PCAs
  print(X[0,:])
  print('*'*100)
  print(Xpca[0,:])

def list_column():
  global df_app
  global columnas_df
  global dicc_fila
  dicc_fila = {}
  columnas_df = list(df_app.columns)

def agregar_fila():
  global df_app  
  for i in columnas_df:
    dicc_fila[i] = input('Ingrese el valor de la columna {} '.format(i))
  df_app = df_app.append(dicc_fila, ignore_index= True)

```

```python
def main():
  global list_inercia, xtrain, xtest, ytrain, ytest , df_app, X, Xpca, y_class, user_data
  df_app = pd.read_csv("https://raw.githubusercontent.com/pedrorotta/PythonIntermedio2022/main/clase5/app.csv")
  labelencoder = LabelEncoder()
  scaler = StandardScaler()
  df_app['UPI Banks'] = labelencoder.fit_transform(df_app['UPI Banks'])
  list_column()
  agregar_fila()
  x_standar = df_app[['Volume (Mn) By Costumers','Volume (Mn)','Value (Cr) by Costumers','Value (Cr)']]
  scaler1 = StandardScaler().fit(x_standar)
  X = scaler1.transform(x_standar)
  X = np.c_[X,df_app['Month'],df_app['UPI Banks']]
  
  list_inercia =[]
  for i in range(1,11):
    k_means = KMeans(n_clusters=i) 
    k_means.fit(X)
    list_inercia.append(k_means.inertia_)
  plt.plot(range(1,11),list_inercia)
  plt.show()
  n_kmeans = int(input("Ingrese el número de clusters a crear: "))
  kmeans(n_kmeans)

  pca(X)
  user_data = Xpca[len(Xpca)-1]
  Xpca_1 = Xpca[:len(Xpca)-1,:]
  y_class = y_class[:len(Xpca)-1]

  xtrain,xtest,ytrain,ytest = tts(Xpca_1,y_class,test_size =.30, random_state= 0)

  print("Elegir qué metodo va a utilizar:")
  print("1. Random Forest")
  print("2. SVM")
  print("3. Regresión Logística")
  print("4. KNN")
  metodo = int(input())


  if metodo == 1:
    RF = rfc(n_estimators=100)
    #Entrenar
    RF.fit(xtrain,ytrain)
    #Predecir
    y_ptrain = RF.predict(xtrain)
    y_ptest = RF.predict(xtest)
    rpta = RF.predict(np.reshape(user_data,(1,2)))
    print(rpta)
    print("Hasta Luego")

  elif metodo ==2:
    #Definir modelo
    svc = SVC(gamma = 'scale')
    #Entrenar
    svc.fit(xtrain,ytrain)
    #Predecir
    y_ptrain = svc.predict(xtrain)
    y_ptest = svc.predict(xtest)
    rpta = svc.predict(np.reshape(user_data,(1,2)))
    print(rpta)
    print("Hasta Luego")

  elif metodo == 3:
    #Definimos el modelo
    m_logit = logit(max_iter=50)
    #ENTRENAMIENTO
    m_logit.fit(xtrain,ytrain)
    #Hallar valores predichos
    y_ptrain = m_logit.predict(xtrain)
    y_ptest = m_logit.predict(xtest)
    rpta = m_logit.predict(np.reshape(user_data,(1,2)))
    print(rpta)
    print("Hasta Luego")

  elif metodo == 4:
    #Modelación y Entrenamiento 
    modelo_knn = knn(n_neighbors=3)
    modelo_knn.fit(xtrain, ytrain)
    yptrain_3 = modelo_knn.predict(xtrain)
    yptest_3 = modelo_knn.predict(xtest)
    rpta = modelo_knn.predict(np.reshape(user_data,(1,2))) 
    print(rpta)
    print("Hasta Luego")

  else:
    print("No conozco ese método")
    exit()

main()
```
